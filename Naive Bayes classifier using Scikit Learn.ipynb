# Naive Bayes classifier using Scikit Learn

__Naive Bayes classifier__

A Naive Bayes classifier is a probabilistic machine learning algorithm based on Bayes' theorem. It is called "naive" because it makes the assumption that all the features in the dataset are independent of each other, given the class label.

Naive Bayes classifiers are commonly used for text classification, spam filtering, and sentiment analysis. They are also commonly used in medical diagnosis and weather forecasting.

__When we can use Naive Bayes classifier?__

Naive Bayes classifiers are commonly used in a variety of applications and scenarios, some examples include:

__Text classification:__ Naive Bayes is particularly well-suited for text classification tasks, such as spam filtering, sentiment analysis, and topic classification.

__Medical diagnosis:__ Naive Bayes classifiers have been used in medical diagnosis to classify diseases based on symptoms and test results.

__Weather forecasting:__ Naive Bayes classifiers have been used in weather forecasting to classify weather patterns based on historical data.

__Image classification:__ Naive Bayes classifiers can be used for image classification tasks by treating each pixel as a binary feature, indicating whether it's black or white.

__Real-time classification:__ Naive Bayes classifiers are computationally efficient and can be used for real-time classification tasks, such as classifying incoming email as spam or not spam.

Naive Bayes classifiers are simple, fast and easy to implement, and they work well in many real-world applications. However, it's important to keep in mind that they make the assumption that the features are independent of each other, which is not always the case and might lead to poor performance.

__When we shouldn't use Naive Bayes classifier?__

Naive Bayes classifiers are simple, fast, and easy to implement, but they may not be the best choice in certain scenarios:

__Data with strong dependencies between features:__ The independence assumption between features of Naive Bayes can lead to poor performance when the features are highly dependent on each other.

__Data with rare features:__ The probabilistic approach of Naive Bayes can lead to poor performance when there are rare features in the data, as the classifier will have little information about the probability of these features.

__Limited data:__ The Naive Bayes classifier assumes that the data follows a specific probability distribution, if the data is limited, it may not be possible to estimate this distribution accurately.

__Large number of features:__ The computational complexity of Naive Bayes increases with the number of features, so it may not be suitable when the number of features is very large.

__High-dimensional continuous data:__ Naive Bayes classifiers are not suitable for high-dimensional continuous data, in this case, other techniques like Neural Networks, Random Forest or SVM, may be more suitable.

__Categorical data:__ Naive Bayes classifiers are not suitable for categorical data, in this case, other techniques like Decision Trees, Random Forest or Neural Networks, may be more suitable.

It's important to keep in mind that, while Naive Bayes has some limitations, it still has many advantages and can be a good choice for many problems. It's important to have a good understanding of the problem, the data, and the technique and also to consider other techniques that might be more suitable for the problem at hand.

```python
# Import the necessary libraries
import numpy as np
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load the iris dataset
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
```

```python
# Create an instance of the GaussianNB class
nb = GaussianNB()
```

```python
# Fit the model to the training data
nb.fit(X, y)
```

```python
# Create a meshgrid of points in the feature space
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
```

```python
# Predict the class labels of the meshgrid points
Z = nb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plot the decision boundaries and the data points
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=ListedColormap(['#AAAAFF','#AAFFAA','#FFFFAA']))
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=ListedColormap(['#AAAAFF','#AAFFAA','#FFFFAA']))
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Naive Bayes classification on the Iris dataset")
plt.show()
```

```python
y_pred = nb.predict(X)
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred, average='macro')
recall = recall_score(y, y_pred, average='macro')
f1 = f1_score(y, y_pred, average='macro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
