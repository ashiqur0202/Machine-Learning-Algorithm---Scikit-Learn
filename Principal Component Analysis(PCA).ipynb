# Principal Component Analysis(PCA)

__Principal Component Analysis (PCA)__

Principal Component Analysis (PCA) is a technique used for dimensionality reduction and feature extraction in machine learning and statistics. The main goal of PCA is to identify patterns in data, by projecting it onto a new set of axes, called principal components, that capture the most variation in the data.

PCA works by finding the eigenvectors of the covariance matrix of the data, which are the directions of maximum variance. The eigenvectors are then used to create a new set of orthogonal axes, called principal components, that capture the most variation in the data. The data can then be projected onto these new axes, effectively reducing the dimensionality of the data.

__When we shouldn't use pca?__

Principal Component Analysis (PCA) is a powerful technique for dimensionality reduction and feature extraction, but it may not be the best choice in certain scenarios:

__Categorical data:__ PCA is not suitable for categorical data, as it is based on finding the eigenvectors of the covariance matrix of the data, which assumes that the data is continuous.

__Preserving the class labels:__ PCA is an unsupervised technique, it does not consider the class labels of the data, and it's not designed to preserve the class separation in the data. In scenarios where it is important to preserve the class separation, other techniques such as Linear Discriminant Analysis (LDA) or Quadratic Discriminant Analysis (QDA) might be more suitable.

__Data with missing values:__ PCA is sensitive to missing data, so it's important to handle missing values before applying PCA.

__Data with small number of samples:__ PCA requires a large number of samples to work effectively, it might not be a good choice when the number of samples is small.

__Interpretability:__ PCA can be less interpretable, as the principal components are linear combinations of the original features, and it might be hard to understand the meaning of the new features.

__Sensitive to scaling:__ PCA is sensitive to scaling, so it's important to scale the data before applying PCA.

It's important to note that, before using PCA, it's important to have a good understanding of the problem, the data, and the technique and also to consider other techniques that might be more suitable for the problem at hand.

```python
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Load the iris dataset
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
# Create a PCA object with 2 components
pca = PCA(n_components=2)
```

```python
# Fit the PCA model to the data
pca.fit(X)
```

```python
# Transform the data onto the first two principal components
X_pca = pca.transform(X)
```

```python
# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
