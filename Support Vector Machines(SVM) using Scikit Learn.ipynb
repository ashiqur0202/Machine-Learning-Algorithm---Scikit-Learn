# Support Vector Machines(SVM) using Scikit Learn

__Support Vector Machines__

Support Vector Machines (SVMs) are a type of supervised machine learning algorithm that can be used for classification or regression tasks. The basic idea behind SVMs is to find the hyperplane in a high-dimensional space that best separates the data points into different classes. A hyperplane is a decision boundary that separates the feature space into different regions, each corresponding to a different class.

SVMs are particularly useful when the data is not linearly separable, which means that it can't be separated into different classes by a single straight line. In these cases, SVMs can use a technique called the kernel trick to transform the data into a higher-dimensional space where it becomes linearly separable. The most popular kernel functions are the linear, polynomial, radial basis function (RBF), and sigmoid kernels.

SVMs are known to be effective in high-dimensional spaces, memory-efficient and versatile. They have been widely used in many applications such as text classification, image classification, bioinformatics and many more.

SVMs are also known for their robustness to overfitting, since they only depend on a subset of the data points, called support vectors, to define the decision boundary.

__When I can use Support Vector Machines?__

Support Vector Machines (SVMs) can be used in a variety of scenarios where the goal is to separate or classify data points into different classes. Some examples of when SVMs can be useful include:

__Binary Classification:__ When the data has two classes, and the goal is to separate them with a linear or non-linear decision boundary.

__Multi-class Classification:__ When the data has more than two classes, SVMs can be used in a one-versus-one or one-versus-all approach to perform multi-class classification.

__Non-Linear classification:__ When the data is not linearly separable, SVMs can be used with a non-linear kernel function to transform the data into a higher-dimensional space where it becomes linearly separable.

__High-dimensional spaces:__ SVMs are particularly effective in high-dimensional spaces, where the number of features is greater than the number of samples.

__Handling outliers and noise:__ SVMs are robust to outliers and noise in the data, as they only depend on a subset of the data points, called support vectors, to define the decision boundary.

__Regression:__ SVMs can also be used for regression tasks by using a different loss function and replacing the hyperplane with a function that predicts the target value.

It's important to note that, before applying SVM, it is necessary to make sure that the data is cleaned and preprocessed, and also have a good understanding of the problem and the data. It's also important to try different kernel functions and parameters to find the best one for the problem at hand.

__When  using Support Vector Machines can be a bad idea?__

While Support Vector Machines (SVMs) are a powerful and versatile algorithm, there are certain scenarios where they may not be the best choice:

__Large datasets:__ SVMs can be computationally expensive when the number of data points is very large, as the algorithm has to solve a quadratic optimization problem. For very large datasets, other algorithms such as Random Forest or Gradient Boosting may be more efficient.

__High dimensionality:__ Although SVMs are effective in high-dimensional spaces, they can become less effective as the number of features becomes much larger than the number of samples. In this case, other algorithms such as Principal Component Analysis (PCA) can be used to reduce the dimensionality of the data.

__Data with missing values:__ SVMs are sensitive to missing data, so it's important to handle missing values before training the model.

__Non-linearly inseparable data:__ SVMs can become less effective when the data is not linearly separable, and the kernel trick is not able to transform the data into a higher-dimensional space where it becomes linearly separable. In this case, other algorithms such as neural networks might be better suited.

__Text or image classification:__ SVMs can be less effective when dealing with large amount of text or image data. In these cases, other algorithms such as convolutional neural networks (CNN) or recurrent neural networks (RNN) are better suited.

It's important to note that, before choosing SVM as a model, it's necessary to perform some exploratory data analysis, feature engineering, and other preprocessing steps to make sure that the data is cleaned and ready for modeling. Also, it's important to evaluate the performance of the model using different evaluation metrics and try different kernel functions and parameters to find the best one for the problem at hand.

```python
# Import the necessary libraries
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load the iris dataset
iris = datasets.load_iris()
```

```python

X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
```

```python

# Create an instance of the SVC class
svm = SVC(kernel='linear', C=1, random_state=0)
```

```python
# Fit the model to the training data
svm.fit(X, y)
```

```python
# Create a meshgrid of points in the feature space
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
```

```python
# Predict the class labels of the meshgrid points
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plot the decision boundaries and the data points
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=ListedColormap(['#AAAAFF','#AAFFAA','#FFFFAA']))
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=ListedColormap(['#AAAAFF','#AAFFAA','#FFFFAA']))
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("SVM classification on the Iris dataset")
plt.show()
```

```python
y_pred = svm.predict(X)
print("Accuracy:", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred, average='macro'))
print("Recall:", recall_score(y, y_pred, average='macro'))
```

__Advantage of Support Vector Machines__

Support Vector Machines (SVMs) are a popular and powerful machine learning algorithm that have several advantages:

__Effective in high-dimensional spaces:__ SVMs are particularly effective in high-dimensional spaces, where the number of features is greater than the number of samples. They can still find a decision boundary even in cases where the data is not linearly separable.

__Robust to overfitting:__ SVMs are robust to overfitting, as they only depend on a subset of the data points, called support vectors, to define the decision boundary. This means that they are less likely to be affected by noise or outliers in the data.

__Versatile:__ SVMs can be used for both classification and regression tasks. They can also handle non-linear decision boundaries using the kernel trick.

__Memory-efficient:__ SVMs are memory-efficient, as they only need to store the support vectors, which are a subset of the data points.

__Provide probability estimates:__ Some implementations of SVM like the Platt Scaling, can provide probability estimates for the class labels. This can be useful in certain applications where the probability of a certain class label is needed.

__Handling imbalanced data:__ SVMs have a built-in regularization parameter that allows handling imbalanced data by setting different penalties for different classes.

It's important to note that, while SVMs have many advantages, it's necessary to have a good understanding of the problem, the data, and the algorithm to be able to select the best parameters, kernel functions and to avoid overfitting.

__Disadvantage of Support Vector Machines__

Support Vector Machines (SVMs) are a powerful and versatile machine learning algorithm, but they also have several disadvantages:

__Computationally expensive:__ SVMs can be computationally expensive when the number of data points is very large. The algorithm has to solve a quadratic optimization problem, which can become computationally intractable as the number of data points increases.

__Sensitive to the choice of kernel function:__ The performance of SVMs heavily depends on the choice of kernel function. Choosing the wrong kernel function can lead to poor performance or overfitting.

__Sensitive to the choice of parameters:__ The performance of SVMs also depends on the choice of parameters, such as the regularization parameter and the kernel parameters. Choosing the wrong values can lead to poor performance or overfitting.

__Not suitable for large datasets:__ SVMs can become slow and memory-intensive when the number of data points is very large. Other algorithms such as Random Forest or Gradient Boosting may be more efficient for large datasets.

__Require data preprocessing:__ SVMs are sensitive to the scale of the features and require data preprocessing, such as normalization or standardization, before being applied to the data.

__Limited interpretability:__ The decision boundary of an SVM model is represented by the support vectors, which are a subset of the data points. In some cases, it might be hard to interpret the decision boundary, especially in high-dimensional spaces.

It's important to note that, while SVMs have some disadvantages, they still have many advantages and can be a good choice for many problems. It's important to have a good understanding of the problem, the data, and the algorithm to be able to select the best parameters, kernel functions and to avoid overfitting.
